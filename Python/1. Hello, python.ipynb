# Numbers and arithmetic in Python

When we meet a var in Python, we can ask Python the data type by using
type() function.

In Python, we have lots of data types.

1.  int : integer

```python
var = 0
type(var) # Here the output is int.
```

1.  float : floating point number

```python
var = 0.0
type(var) # Here the output is float.
```

1.  bool : boolean(True/False)

```python
var = True
type(var) # Here the output is bool.  
```

A natural thing we want to do with numbers is arithmetic. In Python, we
have lots of operators for arithmetic.

a+b addition a-b subtraction a\*b multiplication a/b division a//b floor
division a%b modulus a\*\*b exponentiation -a negation

An intersting observation is that Python gives us 2 types of division.

```python
print(5 / 2) # 2.5
print(6 / 2) # 3.0
# The interesting thing is that the data type is float.

print(5 // 2) # 2
print(6 // 2) # 3
# // operator give us the int type
```

## Builtin functions for working with numbers

min and max function reture the minimum and maximum of their arguments.

```python
print(min(1, 2, 3)) #1
print(max(1, 2, 3)) #3
```

abs function returns the absolute value of it argument.

```python
print(abs(32)) #32
print(abs(-32)) #32
```

int() and float() and convert numbers into corresponding tyoe.

```python
print(float(10)) #10.0
print(int(3.33)) #3
```
